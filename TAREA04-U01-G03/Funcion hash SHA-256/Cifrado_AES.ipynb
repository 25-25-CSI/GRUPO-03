```python
import os
import time
import base64
import csv
import random
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from Crypto.Util.Padding import pad, unpad

def generar_archivo_texto(nombre_archivo, cantidad_palabras):
    palabras_base = [
        "hola", "mundo", "python", "clave", "mensaje", "criptografia", "seguridad", "texto", "cifrado", "descifrado",
        "codigo", "archivo", "datos", "usuario", "sistema", "programa", "funcion", "valor", "cadena", "binario",
        "bloque", "vector", "inicializacion", "algoritmo", "AES", "CBC", "modo", "bytes", "clave_secreta", "operacion",
        "rango", "repetir", "lectura", "escritura", "guardar", "procesar", "tiempo", "medir", "resultado", "verificar",
        "comparar", "igualdad", "correcto", "error", "excepcion", "control", "flujo", "decodificar", "codificar", "base64",
        "llave", "criptograma", "texto_plano", "padding", "relleno", "bloques", "longitud", "seguro", "confidencialidad", "integridad",
        "funcionalidad", "clave_aleatoria", "generar", "encriptar", "descifrar", "desencriptar", "buffer", "contenido", "palabra", "frase",
        "mensaje_secreto", "aleatorio", "entrada", "salida", "original", "convertir", "formato", "bytearray", "hexadecimal", "string",
        "caracter", "simbolo", "operacion_logica", "bit", "bitwise", "bucle", "iteracion", "contador", "condicion", "logica",
        "aucas", "else", "try", "except", "return", "print", "input", "open", "close", "with",
        "write", "read", "append", "list", "dict", "set", "tuple", "import", "from", "papa",
        "campeon", "decode", "encode", "hash", "SHA256", "digest", "modulo", "libreria", "paquete", "modulo_crypto",
        "pip", "instalar", "dependencia", "version", "script", "main", "def", "parametro", "argumento", "tipo",
        "excepcion_manejada", "segundo", "milisegundo", "tiempo_total", "rendimiento", "optimizacion", "costo", "memoria", "uso", "tamaño",
        "procesador", "archivo_grande", "experimento", "medicion", "comparativa", "benchmark", "evaluacion", "texto_largo", "escenario", "prueba",
        "automatizar", "bateria", "ciclo", "test", "unitario", "integracion", "entorno", "colaborativo", "modular", "reutilizable",
        "estructura", "organizacion", "desarrollo", "ingenieria", "informacion", "clave_privada", "clave_publica", "simetrico", "asimetrico", "certificado",
        "firmar", "verificar_firma", "seguridad_digital", "privacidad", "proteccion", "ciberseguridad", "internet", "red", "comunicacion", "protocolo"
    ]

    with open(nombre_archivo, 'w', encoding='utf-8') as f:
        palabras = [random.choice(palabras_base) for _ in range(cantidad_palabras)]
        f.write(" ".join(palabras))

def leer_archivo(ruta):
    inicio = time.time()
    with open(ruta, 'r', encoding='utf-8') as f:
        contenido = f.read()
    fin = time.time()
    tiempo = (fin - inicio) * 1000
    print(f" Tiempo de lectura del archivo: {tiempo:.2f} ms")
    return contenido, tiempo

def encode(key, text):
    cipher = AES.new(key, AES.MODE_CBC)
    text_bytes = cipher.encrypt(pad(text.encode(), AES.block_size))
    iv = base64.b64encode(cipher.iv).decode('utf-8')
    texto_encriptado = base64.b64encode(text_bytes).decode('utf-8')
    return iv, texto_encriptado

def decode(key, iv, text):
    try:
        iv = base64.b64decode(iv)
        text = base64.b64decode(text)
        cipher = AES.new(key, AES.MODE_CBC, iv)
        pt = unpad(cipher.decrypt(text), AES.block_size)
        return pt.decode('utf-8')
    except (ValueError, KeyError) as e:
        return f"Error al descifrar: {str(e)}"

def procesar_archivo(archivo, cantidad_palabras=None):
    if cantidad_palabras is not None and not os.path.exists(archivo):
        generar_archivo_texto(archivo, cantidad_palabras)
        print(f" Archivo generado: {archivo}")

    print(f"\n Procesando archivo: {archivo}")

    texto, tiempo_lectura = leer_archivo(archivo)
    caracteres_entrada = len(texto)
    num_palabras = len(texto.split())

    inicio_clave = time.time()
    key = get_random_bytes(32)
    fin_clave = time.time()
    tiempo_clave = (fin_clave - inicio_clave) * 1000
    print(" Clave generada (base64):", base64.b64encode(key).decode())
    print(f" Tiempo de generación de clave: {tiempo_clave:.2f} ms")

    inicio_cifrado = time.time()
    iv, texto_cifrado = encode(key, texto)
    fin_cifrado = time.time()
    tiempo_cifrado = (fin_cifrado - inicio_cifrado) * 1000
    caracteres_salida = len(texto_cifrado)
    print(" El texto se cifró correctamente.")
    print(f" Tiempo de cifrado: {tiempo_cifrado:.2f} ms")

    inicio_descifrado = time.time()
    texto_descifrado = decode(key, iv, texto_cifrado)
    fin_descifrado = time.time()
    tiempo_descifrado = (fin_descifrado - inicio_descifrado) * 1000
    print(" El texto se descifró correctamente.")
    print(f" Tiempo de descifrado: {tiempo_descifrado:.2f} ms")

    coincide = texto == texto_descifrado
    validacion = "Válido" if coincide else "No válido"
    print(f" Validación: {validacion}")

    return [num_palabras, caracteres_entrada, caracteres_salida, tiempo_lectura, tiempo_clave, tiempo_cifrado, tiempo_descifrado]

if __name__ == "__main__":
    cantidades = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
    resultados = []

    for cantidad in cantidades:
        archivo = f"text_{cantidad}_palabras.txt"
        resultado = procesar_archivo(archivo, cantidad)
        resultados.append(resultado)

    while True:
        archivos_txt = [f for f in os.listdir('.') if f.endswith('.txt')]
        print("\n Archivos disponibles en el directorio actual:")
        for i, nombre in enumerate(archivos_txt):
            print(f"{i+1}. {nombre}")

        seleccion = input("\n Ingresa el número del archivo que deseas procesar (ENTER para salir): ").strip()

        if not seleccion:
            break
        elif seleccion.isdigit() and 1 <= int(seleccion) <= len(archivos_txt):
            archivo_seleccionado = archivos_txt[int(seleccion) - 1]
            resultado_manual = procesar_archivo(archivo_seleccionado)
            resultados.append(resultado_manual)
        else:
            print(" Selección inválida. Intenta nuevamente.")

    with open("resultados_aes.csv", mode='w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            "# Palabras",
            "# Caracteres Entrada",
            "# Caracteres Salida",
            "Tiempo Lectura (ms)",
            "Tiempo Clave (ms)",
            "Tiempo Cifrado (ms)",
            "Tiempo Descifrado (ms)"
        ])
        writer.writerows(resultados)

    print("\n Resultados guardados en 'resultados_aes.csv'")
```
